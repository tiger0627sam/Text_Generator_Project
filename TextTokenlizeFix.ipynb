```python
import os
import json
import re
import datetime
```

```python
dataPath = './Data/Jieba_Backup/'
outPath = './Data/'
files = os.listdir(dataPath)

for i in range(len(files)):
    start = datetime.datetime.now()
    print('file index: ', i, ' Start',  end='')
    with open(dataPath + files[i], 'r', encoding='utf-8') as fileRead:
        tempFileJSON = json.load(fileRead)
        tempContent = tempFileJSON['Context']
        tempContent = [i for i in tempContent if i != '&ldquo']
        tempContent = [i for i in tempContent if i != '&rdquo']
        tempContent = [i for i in tempContent if i != 'ldquo']
        tempContent = [i for i in tempContent if i != 'rdquo']
        tempContent = [i for i in tempContent if i != '?']
        tempContent = [i for i in tempContent if '&' not in i]
    
    with open(outPath + 'Jieba/' + files[i], "w",  encoding="utf-8") as outfile:
        tempJSON = tempFileJSON
        tempJSON['Context'] = tempContent
        jsonDict = json.dumps(tempJSON, indent=2, ensure_ascii=False)
        outfile.write(jsonDict)
    end = datetime.datetime.now()
    print(' Time Cost: ', (end-start), ' Done')
```
