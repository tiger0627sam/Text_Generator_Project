```python
from opencc import OpenCC
from paddlespeech.cli.text.infer import TextExecutor
```

```python
def long_text_auto_punctuation(textPunc, inText):
    # textPunc = TextExecutor()
    result = ''
    i = 0
    flag = False
    while len(inText) > 0:
        i += 10
        if i > len(inText):
            i = len(inText)
            flag = True
        
        tempText = textPunc(text=inText[:i], task='punc')
        lineEnd = tempText.find('。')
        if(lineEnd > 0 and lineEnd != len(tempText)-1):
            result += tempText[:lineEnd+1] + '\n'
            inText = inText[i:]
            i = 0
            
        if(flag):
            result += tempText
            break
    return result
```

```python
# cc = OpenCC('t2s')
txtName = 'Bi-Gram_CKIP'
with open('./Results/'+ txtName + '.txt', 'r') as f:
    tempText = f.read()
text_punc = TextExecutor()
# result = long_text_auto_punctuation(text_punc, cc.convert(tempText))
result = long_text_auto_punctuation(text_punc, tempText)
```

```python
with open ('./Results/'+ txtName +'_AutoPunc.txt', 'w', encoding='utf-8') as outFile:
    outFile.write(result)
```
