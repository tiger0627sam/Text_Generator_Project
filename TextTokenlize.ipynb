```python
from ckiptagger import WS
import jieba
import os
import json
from opencc import OpenCC
from zhon.hanzi import punctuation
import re
import datetime
```

```python
# 切詞用
def strQ2B(ustring):
    """把字串全形轉半形"""
    ss = []
    for s in ustring:
        rstring = ""
        for uchar in s:
            inside_code = ord(uchar)
            if inside_code == 12288:  # 全形空格直接轉換
                inside_code = 32
            elif (inside_code >= 65281 and inside_code <= 65374):  # 全形字元（除空格）根據關係轉化
                inside_code -= 65248
            rstring += chr(inside_code)
        ss.append(rstring)
    return ''.join(ss)
```

```python
cc = OpenCC('s2t')
dataPath = './Data/Origin/'
outPath = './Data/'
files = os.listdir(dataPath)
ws = WS("./data")

for i in range(len(files)):
    start = datetime.datetime.now()
    print('file index: ', i, ' Start',  end='')
    with open(dataPath + files[i], 'r', encoding='utf-8') as fileRead:
        tempFileJSON = json.load(fileRead)
        tempContent = tempFileJSON['Context']
        tempText = re.sub(r"[%s]+" %punctuation, "", tempContent)
        tempText = strQ2B(tempText).replace(' ','').replace("\n", "").replace("\r", "").replace('*','')
        tempText = re.sub(r'[★_-]', "", tempText)
    traChiText = cc.convert(tempText)
    ckipTokenList = ws([traChiText])[0]
    jiebaTokenList = list(jieba.cut(tempText))
    
    with open(outPath + 'Ckip/' + files[i], "w",  encoding="utf-8") as outfile:
        tempJSON = tempFileJSON
        tempJSON['Context'] = ckipTokenList
        jsonDict = json.dumps(tempJSON, indent=2, ensure_ascii=False)
        outfile.write(jsonDict)
    
    with open(outPath + 'Jieba/' + files[i], "w",  encoding="utf-8") as outfile:
        tempJSON = tempFileJSON
        tempJSON['Context'] = jiebaTokenList
        jsonDict = json.dumps(tempJSON, indent=2, ensure_ascii=False)
        outfile.write(jsonDict)
    end = datetime.datetime.now()
    print(' Time Cost: ', (end-start), ' Done')
```
